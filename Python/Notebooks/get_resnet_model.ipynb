# IMPORTS

```python
from tensorflow.keras import Model
from tensorflow.keras.applications.resnet_v2 import ResNet152V2
from tensorflow.keras.models import load_model
```

# Utils

```python
import tensorflow as tf
print(tf.__version__)
```

```python
# Path to various directories
project_root_path = "../"
models_root_path = project_root_path + "Models/"
```

### TO GET INPUT VECTOR X FROM IMAGES WE USE TRANSFER LEARNING THROUGH ResNet152V2 MODEL TRAINED ON 1000 DIFFERENT CLASSES OF IMAGES

```python
model = ResNet152V2(weights = "imagenet")
resnet_model = Model(model.input, model.layers[-2].output)

resnet_model.save(models_root_path + "resnet_model")    # SavedModel format
```

```python
resnet_model.save(models_root_path + "resnet_model.h5")     # HDF5 format
```

```python
# Loading Variables
resnet_model = load_model(models_root_path + "resnet_model")

print(resnet_model.summary())
```

```python
# Loading Variables
resnet_model = load_model(models_root_path + "resnet_model.h5")

print(resnet_model.summary())
```

```python
# To save a lite version of resnet model for node app
converter = tf.lite.TFLiteConverter.from_saved_model(models_root_path + "resnet_model")
converter.allow_custom_ops = True
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS]
tflite_model_quant = converter.convert()

# Save the model.
with open(models_root_path + "/resnet_model.tflite", "wb") as file:
  file.write(tflite_model_quant)
```
